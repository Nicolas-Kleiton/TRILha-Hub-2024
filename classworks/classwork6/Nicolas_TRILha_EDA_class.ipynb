### 1.2 - Limpeza dos dados
---



#### Agora que j√° conseguimos ter uma vis√£o do panor√¢mica dos dados, conseguimos ver que eles est√£o uma porcaria, ent√£o que tal dar uma faxina neles? Remova todos os erros, como dados faltantes, colunas com tipos errados, etc. Mais uma vez, a leitura da documenta√ß√£o do `pandas` √© imprescind√≠vel.

_Sua resposta aqui_

```python
# Para come√ßar vamos verificar os dados faltantes
dataset.isnull().sum()
```

```python
# Vamos ver se h√° elementos duplicados
dataset.duplicated().sum()
```

```python
# Criando uma copia do dataset para corre√ß√£o dos dados
df = dataset.copy()
```

```python
# Come√ßando com "Age": Irei preencher os faltantes com a mediana das idades inv√©s de excluir-las.
df['Age'].fillna(df['Age'].median(), inplace=True)
# Usei a mediana, pois ela mant√©m o equil√≠brio, diferente da m√©dia que poderia ser influenciada por outliers

# Agora que preenchemos as idades, vamos mudar seu tipo para inteiro
df['Age'] = df['Age'].astype(int)
```

```python
# Como a coluna "Cabin" tem bastantes valores nulos optei por remove-la.
df = df.drop('Cabin', axis=1)
```

```python
# Em "Embarked" existe apenas dois dados faltantes, ent√£o podemos excluir as linhas faltantes.
df = df.dropna(subset=['Embarked'])
```

```python
# Agora irei excluir dados que s√£o in√∫teis para a nossa an√°lise:
# PassengerID e ticket ser√£o esses dados que ser√£o removidos
df = dataset.drop(['PassengerId', 'Ticket'], axis=1)
```

```python
# Vamos verificar a exist√™ncia de outliers
# Como podemos ver, a coluna 'Fare' possuem outliers
df[['Fare']].plot.box(figsize=(12, 6), subplots=True, layout=(1, 2), grid=True)
plt.suptitle('Boxplots de Fare', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Vamos calcular o intervalo interquartil (IQR) para cada coluna, come√ßando com 'Fare'
Q1_fare = df['Fare'].quantile(0.25)
Q3_fare = df['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare

# Definindo os limites para identificar outliers
limite_inferior_fare = Q1_fare - 1.5 * IQR_fare
limite_superior_fare = Q3_fare + 1.5 * IQR_fare

# Removendo os outliers
df = df[(df['Fare'] >= limite_inferior_fare) & (df['Fare'] <= limite_superior_fare)]
```

```python
# Vamos verificar como ficou agora:
df.boxplot(column=['Fare'])
plt.show()
```

```python
# Vizualiza√ß√£o dos dados depois da limpeza
df
```

```python
# Verificar novamente os dados faltantes ap√≥s a limpeza:
df.isnull().sum()
```

```python
# Vamos come√ßar exibindo a matriz de correla√ß√£o desses dados

# Primeiramente, para isso, vou converter a coluna 'sex' para n√∫merico
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Selecionando as colunas relevantes para an√°lise de correla√ß√£o
colunas_relevantes = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
matriz_correlacao = df[colunas_relevantes].corr()

# Gr√°fico de correla√ß√£o
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacao, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correla√ß√£o')
plt.show()
```

```python
# Correla√ß√£o entre classe social e a sobreviv√™ncia

# Vamos agrupar os dois tipos de dados para saber quantas pessoas de cada sexo sobreviveram
sex_survived = dataset.groupby('Sex')['Survived'].sum()
sex = sex_survived.index
survived = sex_survived.values
```

```python
# Criando o gr√°fico em barras
plt.bar(sex, survived, color=['lightpink', 'lightblue'])

# Adicionando t√≠tulo e r√≥tulos
plt.title('Sobreviv√™ncia por Sexo')
plt.xlabel('Sexo', fontsize=12)
plt.ylabel('N√∫mero de Sobreviventes', fontsize=12)

# Exibindo o gr√°fico
plt.show()
```

```python
# Correla√ß√£o entre a classe social e a sobreviv√™ncia

# Agrupando os dados
class_survived = df.groupby('Pclass')['Survived'].sum()
class_ = class_survived.index
survived = class_survived.values
```

```python
# Criando o gr√°fico
plt.bar(class_, survived, color=['lightblue', 'lightsalmon', 'lightcoral'])

# Adicionando t√≠tulo e r√≥tulos
plt.title('Sobreviv√™ncia por Classe')
plt.ylabel('N√∫mero de Sobreviventes', fontsize=12)
plt.xticks([1, 2, 3], ['1¬™ Classe', '2¬™ Classe', '3¬™ Classe'])

# Exibindo o gr√°fico
plt.show()
```

```python
# Correla√ß√£o de idade por sobreviv√™ncia com o gr√°fico de violino
sns.violinplot(x='Survived', y='Age', data=df)
plt.title('Distribui√ß√£o de Idade por Sobreviv√™ncia', fontweight="bold")
plt.xlabel('Sobreviveu', fontweight="bold")
plt.ylabel('Idade', fontweight="bold")
plt.show()
```

```python
# Coluna de faixa et√°ria
df.loc[:, 'Faixa_Etaria'] = pd.cut(df['Age'], bins=[0, 12, 18, 60, 100], labels=['Crian√ßa', 'Jovem', 'Adulto', 'Idoso'])

# Coluna de faixa tarifa
df.loc[:, 'Faixa_Tarifa'] = pd.cut(df['Fare'], bins=[0, 10, 30, 100, 1000], labels=['Baixo', 'Medio', 'Alto', 'Muito Alto'])

df
```
